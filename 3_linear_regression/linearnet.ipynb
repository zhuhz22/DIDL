```python
import torch
import random
import numpy as np
num_inputs = 2
num_examples = 1000
true_w = [2, -3.4]
true_b = 4.2
#TODO define features and labels 
features=torch.randn(num_examples,num_inputs)  #x1,x2
labels=list() 
for i in range (len(features)):
    labels.append(features[i][0]*true_w[0]+features[i][1]*true_w[1]+true_b+np.random.normal(0,0.01))
labels=torch.tensor(labels).view(len(features),1)
#TODO end   
```

```python
#TODO define data_iter
def data_iter(batch_size, features, labels):
    for i in range(0,len(features),batch_size):
        features_chunk=features[i:min(i+batch_size,len(features))]
        labels_chunk=labels[i:min(i+batch_size,len(labels))]
        yield features_chunk,labels_chunk
#TODO end

batch_size = 10

#TODO to initialize w_1_2,b
w=torch.tensor(np.random.normal(100,0.1,(1,2)),dtype=torch.float32).view(2,1)  #ndarray
b=torch.tensor(100,dtype=torch.float32)   #float
w.requires_grad=True
b.requires_grad=True
def linreg(X, w, b):  
    return torch.mm(X, w) + b
def squared_loss(y_hat, y):  
    # , , pytorchMSELoss 2
    return ((y_hat - y.view(y_hat.size())) ** 2 / 2).sum()

#TODO define sgd
def sgd(params, lr, batch_size):
    for par in params:
        par.data-=(lr/batch_size)*par.grad
#TODO end

lr = 0.03
num_epochs = 3000
net = linreg
loss = squared_loss

#TODO to train
for epoch in range(num_epochs):
    for feature_chunk,label_chunk in data_iter(batch_size,features,labels):
        l=loss(linreg(feature_chunk,w,b),label_chunk)
        l.backward()
        sgd([w,b],lr,batch_size)
        w.grad.data.zero_()
        b.grad.data.zero_()
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, train_l.mean().item()))
#TODO end  
    
print(true_w, '\n', w)
print(true_b, '\n', b)
```
